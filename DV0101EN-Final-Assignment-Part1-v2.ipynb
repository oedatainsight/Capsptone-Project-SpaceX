````python
import pandas as pd
import matplotlib.pyplot as plt

# Aggregate sales by year and vehicle type
sales_by_type = df.groupby(['Year', 'Vehicle_Type'])['Automobile_Sales'].sum().reset_index()

# Pivot the data to get each vehicle type as a separate column
sales_pivot = sales_by_type.pivot(index='Year', columns='Vehicle_Type', values='Automobile_Sales')

# Plot the data
plt.figure(figsize=(12, 8))
for vehicle_type in sales_pivot.columns:
    plt.plot(sales_pivot.index, sales_pivot[vehicle_type], marker='o', label=vehicle_type)

# Add labels, legend, and title
plt.title('Yearly Automobile Sales by Vehicle Type')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.legend(title='Vehicle Type')
plt.grid()

# Show the plot
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
import matplotlib.pyplot as plt

# Example preparation of data
# Assuming df has columns: 'Season', 'Automobile_Sales', and 'Impact'
season_data = df.groupby('Seasonality_Weight').agg({'Automobile_Sales': 'sum', 'Price': 'mean'}).reset_index()

# Define colors for each season
colors = {'Winter': 'blue', 'Spring': 'green', 'Summer': 'orange', 'Fall': 'brown'}

# Create bubble plot
plt.figure(figsize=(100, 60))
for i, row in season_data.iterrows():
    plt.scatter(
        row['Seasonality_Weight'],
        row['Automobile_Sales'],
        s=row['Price'] * 2,  # Scale bubble size
        color=colors.get(row['Seasonality_Weight'], 'gray'),
        alpha=0.8,
        label=row['Seasonality_Weight']
    )

# Add labels and title
plt.title("Impact of Seasonality on Automobile Sales")
plt.xlabel("Season")
plt.ylabel("Total Sales")
plt.legend(title="Season")
plt.grid()

# Show the plot
plt.show()
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
import matplotlib.pyplot as plt

# Filter data for recession periods
recession_data = df[df['Recession'] == 1]

# Group by relevant columns to calculate averages and totals
recession_summary = recession_data.groupby('Vehicle_Type').agg({
    'Price': 'mean',  # Average price during recessions
    'Automobile_Sales': 'sum'  # Total sales during recessions
}).reset_index()

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(
    recession_summary['Price'],
    recession_summary['Automobile_Sales'],
    color='blue',
    alpha=0.7
)

# Add labels and title
plt.title("Correlation Between Average Vehicle Price and Sales Volume During Recessions")
plt.xlabel("Average Vehicle Price (USD)")
plt.ylabel("Sales Volume")
plt.grid()

# Show the plot
plt.show()
````

````python
import dash
from dash import html, dcc
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Sample dataset
data = {
    'Year': [2015, 2016, 2017, 2018],
    'Recession': [1, 0, 1, 0],
    'Automobile_Sales': [1000, 1500, 800, 1300],
    'Unemployment_Rate': [5.5, 4.7, 6.0, 3.9],
    'Vehicle_Type': ['SUV', 'Sedan', 'Truck', 'Coupe']
}
df = pd.DataFrame(data)

# Initialize the Dash app
app = dash.Dash(__name__)

# Set a meaningful title
app.title = "Automobile Sales Analysis Dashboard"

# Define the layout
app.layout = html.Div([
    html.H1("Automobile Sales Analysis Dashboard", style={'textAlign': 'center'}),
    html.P("Explore insights on vehicle sales, trends, and economic impacts.", style={'textAlign': 'center'}),
    
    # Add drop-downs with appropriate titles and options
    html.Div([
        html.Label("Select a Vehicle Type:"),
        dcc.Dropdown(
            id='vehicle-type-dropdown',
            options=[
                {'label': vehicle, 'value': vehicle} for vehicle in df['Vehicle_Type'].unique()
            ],
            value='SUV',  # Default value
            placeholder="Select a vehicle type",
            style={'width': '50%'}
        )
    ], style={'margin': '20px'}),

    html.Div([
        html.Label("Select a Year:"),
        dcc.Dropdown(
            id='year-dropdown',
            options=[
                {'label': str(year), 'value': year} for year in df['Year'].unique()
            ],
            value=2015,  # Default value
            placeholder="Select a year",
            style={'width': '50%'}
        )
    ], style={'margin': '20px'}),

    # Output display division
    html.Div(
        id='output-display',  # Set an ID for the output division
        className='output-class',  # Add a class for styling
        children="Output for appropriate display.",  # Placeholder text
        style={'margin': '20px', 'padding': '10px', 'border': '1px solid #ccc', 'borderRadius': '5px'}
    ),

    # Graphs for Recession Report Statistics
    html.Div([
        html.H3("Recession Report Statistics", style={'textAlign': 'center'}),
        dcc.Graph(id='recession-bar-chart'),
        dcc.Graph(id='recession-line-chart')
    ])
])

# Define the callback to update the output container
@app.callback(
    [
        Output('output-display', 'children'),
        Output('recession-bar-chart', 'figure'),
        Output('recession-line-chart', 'figure')
    ],
    [
        Input('vehicle-type-dropdown', 'value'),
        Input('year-dropdown', 'value')
    ]
)
def update_output(selected_vehicle, selected_year):
    # Update text output
    output_text = f"Selected Vehicle Type: {selected_vehicle}, Selected Year: {selected_year}"

    # Filter data based on selected inputs
    filtered_data = df[(df['Vehicle_Type'] == selected_vehicle) & (df['Year'] == selected_year)]

    # Bar chart for automobile sales
    bar_chart = px.bar(
        df[df['Recession'] == 1],  # Only recession data
        x='Year',
        y='Automobile_Sales',
        color='Vehicle_Type',
        title="Automobile Sales During Recession"
    )

    # Line chart for unemployment rate
    line_chart = px.line(
        df[df['Recession'] == 1],  # Only recession data
        x='Year',
        y='Unemployment_Rate',
        title="Unemployment Rate During Recession",
        markers=True
    )

    return output_text, bar_chart, line_chart


# Run the Dash app
if __name__ == "__main__":
    app.run_server(debug=True)
````
